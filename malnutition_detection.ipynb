```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import zipfile
import os

os.chdir("/content")

zip_files = ["archive (5).zip", "archive (6).zip", "archive (7).zip"]

# Extract all ZIP files
for zip_file in zip_files:
    if os.path.exists(zip_file):
        with zipfile.ZipFile(zip_file, 'r') as zip_ref:
            zip_ref.extractall("/content/")
    else:
        print(f"❌ {zip_file} not found. Please upload the file!")

print("✅ All files extracted successfully!")

nfhs5_df = pd.read_csv("/content/datafile.csv", encoding="utf-8")
income_df = pd.read_csv("/content/data.csv", encoding="utf-8")
food_price_df = pd.read_csv("/content/Price_Agriculture_commodities_Week.csv", encoding="utf-8")

nfhs5_df.columns = nfhs5_df.columns.str.strip()
income_df.columns = income_df.columns.str.strip()
food_price_df.columns = food_price_df.columns.str.strip()

nfhs5_clean = nfhs5_df[["District Names", "State/UT",
                         "Children under 5 years who are stunted (height-for-age)18 (%)",
                         "Children under 5 years who are wasted (weight-for-height)18 (%)",
                         "Children under 5 years who are severely wasted (weight-for-height)19 (%)",
                         "Children under 5 years who are underweight (weight-for-age)18 (%)"]].copy()

nfhs5_clean.columns = ["District", "State", "Stunted (%)", "Wasted (%)", "Severely Wasted (%)", "Underweight (%)"]

income_clean = income_df[["State Name", "NSDP Per Capita (Nominal)2019-20"]].copy()
income_clean.columns = ["State", "Per Capita Income (2019-20)"]

income_clean["Per Capita Income (2019-20)"] = (
    income_clean["Per Capita Income (2019-20)"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
)

income_clean["Per Capita Income (2019-20)"] = pd.to_numeric(income_clean["Per Capita Income (2019-20)"], errors='coerce')

food_price_clean = food_price_df[["State", "District", "Commodity", "Modal Price"]].copy()
food_price_clean["Modal Price"] = pd.to_numeric(food_price_clean["Modal Price"], errors="coerce")

nfhs5_clean["State"] = nfhs5_clean["State"].str.strip().str.title()
income_clean["State"] = income_clean["State"].str.strip().str.title()
food_price_clean["State"] = food_price_clean["State"].str.strip().str.title()

nfhs5_clean["District"] = nfhs5_clean["District"].str.strip().str.title()
food_price_clean["District"] = food_price_clean["District"].str.strip().str.title()

state_corrections = {
    "Maharastra": "Maharashtra",
    "Jammu And Kashmir": "Jammu & Kashmir",
    "Nct Of Delhi": "Delhi",
    "Dadra And Nagar Haveli & Daman And Diu": "Dadra and Nagar Haveli and Daman and Diu",
    "Pondicherry": "Puducherry",
    "Chattisgarh": "Chhattisgarh",
    "Uttrakhand": "Uttarakhand"
}

nfhs5_clean["State"] = nfhs5_clean["State"].replace(state_corrections)
income_clean["State"] = income_clean["State"].replace(state_corrections)
food_price_clean["State"] = food_price_clean["State"].replace(state_corrections)

print("Unique States in nfhs5_clean:", nfhs5_clean["State"].unique())
print("Unique States in income_clean:", income_clean["State"].unique())
print("Unique States in food_price_clean:", food_price_clean["State"].unique())

merged_df = nfhs5_clean.merge(income_clean, on="State", how="outer")
merged_df = merged_df.merge(food_price_clean, on=["State", "District"], how="outer")

for col in ["Stunted (%)", "Wasted (%)", "Severely Wasted (%)", "Underweight (%)", "Modal Price", "Per Capita Income (2019-20)"]:
    merged_df[col] = pd.to_numeric(merged_df[col], errors="coerce")

merged_df.dropna(subset=["Per Capita Income (2019-20)", "Modal Price", "Stunted (%)", "Wasted (%)", "Severely Wasted (%)", "Underweight (%)"], how="all", inplace=True)

print("Number of rows after merging:", len(merged_df))
print(merged_df.head())

print(merged_df[["Per Capita Income (2019-20)", "Modal Price", "Stunted (%)", "Wasted (%)", "Severely Wasted (%)", "Underweight (%)"]].isna().sum())

correlation_matrix = merged_df.dropna()[["Per Capita Income (2019-20)", "Modal Price", "Stunted (%)", "Wasted (%)", "Severely Wasted (%)", "Underweight (%)"]].corr()

plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Income, Food Prices, and Malnutrition")
plt.show()

print("✅ Final Cleaned Dataset:")
print(merged_df.head())
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the cleaned dataset (Ensure it has necessary columns)
df = merged_df.copy()  # Assuming 'merged_df' is the preprocessed DataFrame

# Drop missing values for accurate regression
df.dropna(subset=["Per Capita Income (2019-20)", "Modal Price", "Stunted (%)", "Wasted (%)", "Severely Wasted (%)", "Underweight (%)"], inplace=True)

# Independent variables
X = df[["Per Capita Income (2019-20)", "Modal Price"]]  # Predictors
y_stunted = df["Stunted (%)"]
y_wasted = df["Wasted (%)"]
y_underweight = df["Underweight (%)"]

# Split into training & testing sets
X_train, X_test, y_train_stunted, y_test_stunted = train_test_split(X, y_stunted, test_size=0.2, random_state=42)
_, _, y_train_wasted, y_test_wasted = train_test_split(X, y_wasted, test_size=0.2, random_state=42)
_, _, y_train_underweight, y_test_underweight = train_test_split(X, y_underweight, test_size=0.2, random_state=42)

# Initialize and train linear regression models
reg_stunted = LinearRegression().fit(X_train, y_train_stunted)
reg_wasted = LinearRegression().fit(X_train, y_train_wasted)
reg_underweight = LinearRegression().fit(X_train, y_train_underweight)

# Predictions
y_pred_stunted = reg_stunted.predict(X_test)
y_pred_wasted = reg_wasted.predict(X_test)
y_pred_underweight = reg_underweight.predict(X_test)

# Model Performance
def evaluate_model(y_test, y_pred, label):
    print(f"\n🔹 {label} Regression Results:")
    print(f"  - R² Score: {r2_score(y_test, y_pred):.4f}")
    print(f"  - Mean Squared Error: {mean_squared_error(y_test, y_pred):.4f}")

evaluate_model(y_test_stunted, y_pred_stunted, "Stunted")
evaluate_model(y_test_wasted, y_pred_wasted, "Wasted")
evaluate_model(y_test_underweight, y_pred_underweight, "Underweight")

# Multivariate Regression using Statsmodels
X_const = sm.add_constant(X)  # Add constant for intercept
model = sm.OLS(df["Stunted (%)"], X_const).fit()  # Model for 'Stunted (%)'

# Summary of regression results
print("\n🔹 Multivariate Regression Summary for 'Stunted (%)':")
print(model.summary())

# 🔥 Data Visualization
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["Per Capita Income (2019-20)"], y=df["Stunted (%)"], color='blue', label="Stunted")
sns.scatterplot(x=df["Per Capita Income (2019-20)"], y=df["Wasted (%)"], color='red', label="Wasted")
sns.scatterplot(x=df["Per Capita Income (2019-20)"], y=df["Underweight (%)"], color='green', label="Underweight")
plt.title("Malnutrition vs Per Capita Income")
plt.xlabel("Per Capita Income (₹)")
plt.ylabel("Malnutrition Rates (%)")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 1️⃣ Scatter Plot: Income vs Malnutrition (Highlighting Anomalies)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_df, x="Per Capita Income (2019-20)", y="Underweight (%)", hue="State", palette="viridis", alpha=0.7)
plt.axhline(y=merged_df["Underweight (%)"].median(), color='r', linestyle='--', label="Median Malnutrition")
plt.axvline(x=merged_df["Per Capita Income (2019-20)"].median(), color='b', linestyle='--', label="Median Income")
plt.xlabel("Per Capita Income (₹)")
plt.ylabel("Underweight (%)")
plt.title("Income vs Malnutrition (Inequality & Outliers)")
plt.legend()
plt.show()

# 2️⃣ Identifying States/Districts with Higher Malnutrition Despite High Income
high_income_threshold = merged_df["Per Capita Income (2019-20)"].quantile(0.75)
high_malnutrition_threshold = merged_df["Underweight (%)"].quantile(0.75)

high_income_high_malnutrition = merged_df[
    (merged_df["Per Capita Income (2019-20)"] > high_income_threshold) &
    (merged_df["Underweight (%)"] > high_malnutrition_threshold)
]

print("⚠️ States/Districts with HIGH Income but HIGH Malnutrition (Possible Inequality):")
print(high_income_high_malnutrition[["State", "District", "Per Capita Income (2019-20)", "Underweight (%)"]])

# 3️⃣ Identifying States/Districts with Lower Malnutrition Despite Low Income (Policy Success)
low_income_threshold = merged_df["Per Capita Income (2019-20)"].quantile(0.25)
low_malnutrition_threshold = merged_df["Underweight (%)"].quantile(0.25)

low_income_low_malnutrition = merged_df[
    (merged_df["Per Capita Income (2019-20)"] < low_income_threshold) &
    (merged_df["Underweight (%)"] < low_malnutrition_threshold)
]

print("✅ States/Districts with LOW Income but LOW Malnutrition (Possible Policy Success):")
print(low_income_low_malnutrition[["State", "District", "Per Capita Income (2019-20)", "Underweight (%)"]])

# 4️⃣ Correlation Analysis: Food Prices vs Malnutrition
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_df, x="Modal Price", y="Underweight (%)", hue="State", palette="coolwarm", alpha=0.7)
plt.xlabel("Food Price (₹)")
plt.ylabel("Underweight (%)")
plt.title("Food Prices vs Malnutrition (Correlation Analysis)")
plt.legend()
plt.show()

# 5️⃣ Calculating Correlation Between Food Prices & Malnutrition
correlation_food_malnutrition = merged_df[["Modal Price", "Underweight (%)"]].corr().iloc[0, 1]
print(f"📉 Correlation between Food Prices & Malnutrition: {correlation_food_malnutrition:.2f}")

# Interpretation of Correlation
if correlation_food_malnutrition > 0:
    print("🔴 Higher food prices are associated with higher malnutrition rates.")
elif correlation_food_malnutrition < 0:
    print("🟢 Higher food prices are associated with lower malnutrition rates (possibly due to better food quality).")
else:
    print("⚪ No significant correlation found between food prices and malnutrition.")
```
